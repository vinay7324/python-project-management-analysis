## Task 1: Create DataFrames and Save as CSV

```

# Import required libraries
# This line performs the required operation as part of the current task
import pandas as pd
# This line performs the required operation as part of the current task
import numpy as np

# Creating Employee DataFrame
# This line performs the required operation as part of the current task
employee_df = pd.DataFrame({
# This line performs the required operation as part of the current task
    "ID": ["A001","A002","A003","A004","A005"],
# This line performs the required operation as part of the current task
    "Name": ["John Alter","Alice Luxumberg","Tom Sabestine","Nina Adgra","Amy Johny"],
# This line performs the required operation as part of the current task
    "Gender": ["M","F","M","F","F"],
# This line performs the required operation as part of the current task
    "City": ["Paris","London","Berlin","Newyork","Madrid"],
# This line performs the required operation as part of the current task
    "Age": [25,27,29,31,30]
# This line performs the required operation as part of the current task
})

# Creating Seniority Level DataFrame
# This line performs the required operation as part of the current task
seniority_df = pd.DataFrame({
# This line performs the required operation as part of the current task
    "ID": ["A001","A002","A003","A004","A005"],
# This line performs the required operation as part of the current task
    "Designation": [2,2,3,2,3]
# This line performs the required operation as part of the current task
})

# Creating Project DataFrame
# This line performs the required operation as part of the current task
project_df = pd.DataFrame({
# This line performs the required operation as part of the current task
    "ID": ["A001","A002","A003","A004","A005","A002","A005","A003","A001","A003","A001","A004","A004","A005"],
# This line performs the required operation as part of the current task
    "Project": [f"Project {i}" for i in range(1,15)],
# This line performs the required operation as part of the current task
    "Cost": [1002000,2000000,4500000,5500000,np.nan,680000,400000,350000,np.nan,300000,2000000,1000000,3000000,200000],
# This line performs the required operation as part of the current task
    "Status": ["Finished","Ongoing","Finished","Ongoing","Finished","Failed","Finished","Failed","Ongoing","Finished","Failed","Ongoing","Finished","Finished"]
# This line performs the required operation as part of the current task
})

# Saving DataFrames as CSV files
# This line performs the required operation as part of the current task
employee_df.to_csv("employee.csv", index=False)
# This line performs the required operation as part of the current task
seniority_df.to_csv("seniority.csv", index=False)
# This line performs the required operation as part of the current task
project_df.to_csv("project.csv", index=False)
```

## Task 2: Fill Missing Cost Values Using FOR Loop

```

# Reading project CSV file
# This line performs the required operation as part of the current task
project_df = pd.read_csv("project.csv")

# Calculating running average using FOR loop
# This line performs the required operation as part of the current task
running_sum = 0
# This line performs the required operation as part of the current task
count = 0

# This line performs the required operation as part of the current task
for i in range(len(project_df)):
# This line performs the required operation as part of the current task
    if not pd.isna(project_df.loc[i, "Cost"]):
# This line performs the required operation as part of the current task
        running_sum += project_df.loc[i, "Cost"]
# This line performs the required operation as part of the current task
        count += 1
# This line performs the required operation as part of the current task
    else:
# This line performs the required operation as part of the current task
        project_df.loc[i, "Cost"] = running_sum / count

# Saving updated project file
# This line performs the required operation as part of the current task
project_df.to_csv("project.csv", index=False)
```

## Task 3: Split Name Column

```

# Reading employee CSV file
# This line performs the required operation as part of the current task
employee_df = pd.read_csv("employee.csv")

# Splitting Name into First and Last Name
# This line performs the required operation as part of the current task
employee_df[["First Name", "Last Name"]] = employee_df["Name"].str.split(" ", expand=True)

# Dropping old Name column
# This line performs the required operation as part of the current task
employee_df.drop(columns=["Name"], inplace=True)

# Saving updated employee file
# This line performs the required operation as part of the current task
employee_df.to_csv("employee.csv", index=False)
```

## Task 4: Merge All DataFrames

```

# Reading all CSV files
# This line performs the required operation as part of the current task
employee_df = pd.read_csv("employee.csv")
# This line performs the required operation as part of the current task
seniority_df = pd.read_csv("seniority.csv")
# This line performs the required operation as part of the current task
project_df = pd.read_csv("project.csv")

# Merging DataFrames
# This line performs the required operation as part of the current task
final_df = employee_df.merge(seniority_df, on="ID").merge(project_df, on="ID")

# Displaying Final DataFrame
# This line performs the required operation as part of the current task
final_df
```

## Task 5: Add Bonus Column

```

# Adding Bonus column (5% for Finished projects only)
# This line performs the required operation as part of the current task
final_df["Bonus"] = 0

# This line performs the required operation as part of the current task
for i in range(len(final_df)):
# This line performs the required operation as part of the current task
    if final_df.loc[i, "Status"] == "Finished":
# This line performs the required operation as part of the current task
        final_df.loc[i, "Bonus"] = final_df.loc[i, "Cost"] * 0.05

# This line performs the required operation as part of the current task
final_df
```

## Task 6: Demotion and Deletion

```

# Demoting designation for Failed projects
# This line performs the required operation as part of the current task
for i in range(len(final_df)):
# This line performs the required operation as part of the current task
    if final_df.loc[i, "Status"] == "Failed":
# This line performs the required operation as part of the current task
        final_df.loc[i, "Designation"] += 1

# Deleting employees with designation level greater than 4
# This line performs the required operation as part of the current task
final_df = final_df[final_df["Designation"] <= 4]

# This line performs the required operation as part of the current task
final_df
```

## Task 7: Add Mr./Mrs. and Drop Gender

```

# Adding title based on Gender
# This line performs the required operation as part of the current task
for i in range(len(final_df)):
# This line performs the required operation as part of the current task
    if final_df.loc[i, "Gender"] == "M":
# This line performs the required operation as part of the current task
        final_df.loc[i, "First Name"] = "Mr. " + final_df.loc[i, "First Name"]
# This line performs the required operation as part of the current task
    else:
# This line performs the required operation as part of the current task
        final_df.loc[i, "First Name"] = "Mrs. " + final_df.loc[i, "First Name"]

# Dropping Gender column
# This line performs the required operation as part of the current task
final_df.drop(columns=["Gender"], inplace=True)

# This line performs the required operation as part of the current task
final_df
```

## Task 8: Promote Based on Age Using IF Condition

```

# Promoting designation for employees older than 29
# This line performs the required operation as part of the current task
for i in range(len(final_df)):
# This line performs the required operation as part of the current task
    if final_df.loc[i, "Age"] > 29:
# This line performs the required operation as part of the current task
        final_df.loc[i, "Designation"] -= 1

# This line performs the required operation as part of the current task
final_df
```

## Task 9: Total Project Cost Per Employee

```

# Grouping and summing project costs
# This line performs the required operation as part of the current task
total_cost_df = final_df.groupby(["ID", "First Name"])["Cost"].sum().reset_index()

# Renaming columns
# This line performs the required operation as part of the current task
total_cost_df.columns = ["ID", "First Name", "Total cost"]

# This line performs the required operation as part of the current task
total_cost_df
```
